# Day 17 – Model Deployment Basics for Machine Learning

## Introduction
Model deployment is the process of making a trained Machine Learning model available for real-world use.
After training and evaluating a model, it must be deployed so that users or applications can use it.

Deployment connects Machine Learning models with real-world systems.

---

## Why Model Deployment is Important
A Machine Learning model is useful only when it is used in real applications.

Deployment helps in:
- Making predictions in real time
- Integrating ML with web or mobile applications
- Automating decision-making processes
- Delivering value from ML models

---

## Common Ways to Deploy ML Models
- Web applications (Flask, Django)
- APIs (REST API)
- Cloud platforms (AWS, Azure, GCP)
- Batch processing systems

---

## Steps in Model Deployment
1. Train and save the model  
2. Load the model in an application  
3. Create an interface (API or UI)  
4. Take input data  
5. Return predictions  

---

## Model Serialization
Before deployment, models are saved using serialization.

Common methods:
- Pickle
- Joblib

This allows the model to be reused without retraining.

---

## Challenges in Deployment
- Handling real-time data
- Scalability
- Model updates
- Performance optimization

---

## Conclusion
Model deployment is the final step in the Machine Learning lifecycle.
It transforms a trained model into a usable application that solves real-world problems.

### Traing And Saving A Model

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

# Sample data
data = {
    "StudyHours": [1, 2, 3, 4, 5],
    "Marks": [40, 50, 65, 75, 90]
}

df = pd.DataFrame(data)

X = df[["StudyHours"]]
y = df["Marks"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Save model
joblib.dump(model, "model.pkl")
```

### Loading the saved model

```python
import joblib

model = joblib.load("model.pkl")
```


### Making a prediction

```python
new_data = [[6]]
prediction = model.predict(new_data)

print("Predicted Marks:", prediction)
```

#### Output: Predicted Marks: [105.]
