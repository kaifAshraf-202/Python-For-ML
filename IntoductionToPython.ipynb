 ## Operators ML Foundations

To become a Machine Learning Engineer, it is important to understand how Python works with data.

## 1. Operators in Python (Machine Learning Perspective)

Operators are used to perform calculations, comparisons, and logical decisions.
Almost every machine learning algorithm relies on these operations.

## 1.1 Arithmetic Operators

Arithmetic operators are used for mathematical calculations.



```python
a = 10
b = 3

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
print(a % b)   # Modulus
print(a ** b)  # Power
```

### Why this matters in ML:

Used in loss functions

Used in gradient calculations

Used in statistical formulas

Used in feature scaling

### 1.2 Assignment Operators

Assignment operators update the value of variables.

```python
x = 10
print(x)
x += 5
print(x)
x *= 2
print(x)
```

Why this matters in ML:

Updating model weights

Iterative training processes

Loop-based computations

### 1.3 Comparison Operators

Comparison operators compare two values and return True or False.

```python
x = 50
y = 100

print(x == y)
print(x != y)
print(x < y)
print(x >= y)
```

Why this matters in ML:

Threshold checks

Model performance evaluation

Conditional logic in algorithms

### 1.4 Logical Operators

Logical operators combine multiple conditions.

```python
age = 25
income = 50000

print(age > 18 and income > 30000)
```

Why this matters in ML:

Data filtering

Rule-based decisions

Feature selection logic

Control Flow Statements (Decision Making & Loops)

Machine Learning is not only about data and formulas.
A machine learning program must make decisions and repeat tasks many times.
This is done using control flow statements.

Control flow decides which code will run, when it will run, and how many times it will run.

## 1. Decision Making in Python (if, elif, else)

Decision-making statements allow the program to choose actions based on conditions.

### 1.1 if Statement

The if statement executes code only when a condition is true.

```python
age = 20

if age >= 18:
    print("Eligible for voting")
```

#### ML Usage:

Checking data conditions

Applying rules

Validating inputs

### 1.2 if-else Statement

Used when there are two possible outcomes.

accuracy = 85

if accuracy >= 80:
    print("Model performance is good")
else:
    print("Model needs improvement")


### ML Usage:

Model evaluation

Threshold-based decisions

### 1.3 if-elif-else Statement

Used when there are multiple conditions.

```python
score = 72

if score >= 90:
    print("Excellent")
elif score >= 75:
    print("Good")
else:
    print("Needs improvement")
```

#### ML Usage:

Categorizing results

Performance grading

Decision trees logic (basic idea)

## Day 5 – Functions in Python for Machine Learning

### THEORY
What is a Function?

A function is a block of reusable code that performs a specific task.
Instead of writing the same code again and again, we write it once inside a function and reuse it whenever needed.

In Machine Learning, functions help in:

Data preprocessing

Feature engineering

Model training

Evaluation

Code organization

Functions make ML code clean, modular, and easy to maintain.

Why Functions are Important for Machine Learning

Machine Learning projects are large and complex.
Functions help by:

Breaking a big problem into small parts

Improving code readability

Reducing repetition

Making debugging easier

Supporting reusable ML pipelines

Almost every ML library internally uses functions.

Types of Functions in Python

Built-in Functions
Examples: print(), len(), type(), sum()

User-defined Functions
Functions created by the programmer.

### Function Syntax

```python
def function_name(parameters):
    code
    return value
# def keyword is used to define a function

# Parameters are inputs to the function

# return sends output back
```

```python
# 1. Simple Function Example
def greet():
    print("Welcome to Machine Learning")


# Calling the function:

greet()
```

 # 2. Function with Parameters

```python

def add(a, b):
    return a + b

result = add(10, 5)
print(result)


# ML usage:
# Used in calculations like loss, accuracy, or averages.
```


## 3. Function for Data Processing

```python
def square(number):
    return number * number

print(square(4))


# ML usage:
# Feature transformation and mathematical operations.
```


## 4. Function with Multiple Parameters




```python

def calculate_mean(a, b, c):
    return (a + b + c) / 3

mean_value = calculate_mean(10, 20, 30)
print(mean_value)


# ML usage:
# Statistical calculations on data.
```

## 5. Function Returning Multiple Values


```python
def min_max(numbers):
    return min(numbers), max(numbers)

values = [10, 20, 5, 40]
minimum, maximum = min_max(values)

print(minimum, maximum)


# ML usage:
# Used in data normalization and analysis.
```


## 6. Function for Model Evaluation Logic (Basic)


```python
def check_accuracy(acc):
    if acc >= 80:
        return "Good Model"
    else:
        return "Needs Improvement"

print(check_accuracy(85))


# ML usage:
# Evaluating model performance.

# Why Day 5 is Critical for ML Engineers

# ML pipelines are built using functions

# Reusability saves time and effort

# Clean structure helps in collaboration

# Functions are the base of ML libraries
```
