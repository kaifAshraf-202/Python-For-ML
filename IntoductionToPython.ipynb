## 1. Operators in Python (Machine Learning Perspective)

Operators are used to perform calculations, comparisons, and logical decisions.
Almost every machine learning algorithm relies on these operations.

## 1.1 Arithmetic Operators

Arithmetic operators are used for mathematical calculations.



```python
a = 10
b = 3

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
print(a % b)   # Modulus
print(a ** b)  # Power
```

### Why this matters in ML:

Used in loss functions

Used in gradient calculations

Used in statistical formulas

Used in feature scaling

### 1.2 Assignment Operators

Assignment operators update the value of variables.

```python
x = 10
print(x)
x += 5
print(x)
x *= 2
print(x)
```

Why this matters in ML:

Updating model weights

Iterative training processes

Loop-based computations

### 1.3 Comparison Operators

Comparison operators compare two values and return True or False.

```python
x = 50
y = 100

print(x == y)
print(x != y)
print(x < y)
print(x >= y)
```

Why this matters in ML:

Threshold checks

Model performance evaluation

Conditional logic in algorithms

### 1.4 Logical Operators

Logical operators combine multiple conditions.

```python
age = 25
income = 50000

print(age > 18 and income > 30000)
```

Why this matters in ML:

Data filtering

Rule-based decisions

Feature selection logic
