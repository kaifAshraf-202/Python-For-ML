### 1.3 Comparison Operators

### THEORY

Instead of writing the same code again and again, we write it once inside a function and reuse it whenever needed.

In Machine Learning, functions help in:

Feature engineering

Evaluation

Functions make ML code clean, modular, and easy to maintain.

Why Functions are Important for Machine Learning

Machine Learning projects are large and complex.
Functions help by:

Breaking a big problem into small parts

Improving code readability

Reducing repetition

Making debugging easier

Supporting reusable ML pipelines

Almost every ML library internally uses functions.

Types of Functions in Python

Built-in Functions
Examples: print(), len(), type(), sum()

User-defined Functions
Functions created by the programmer.

### Function Syntax

```python
def function_name(parameters):
    code
    return value
# def keyword is used to define a function

# Parameters are inputs to the function

# return sends output back
```

```python
# 1. Simple Function Example
def greet():
    print("Welcome to Machine Learning")


# Calling the function:

greet()
```

 # 2. Function with Parameters

```python

def add(a, b):
    return a + b

result = add(10, 5)
print(result)


# ML usage:
# Used in calculations like loss, accuracy, or averages.
```


## 3. Function for Data Processing

```python
def square(number):
    return number * number

print(square(4))


# ML usage:
# Feature transformation and mathematical operations.
```


## 4. Function with Multiple Parameters




```python

def calculate_mean(a, b, c):
    return (a + b + c) / 3

mean_value = calculate_mean(10, 20, 30)
print(mean_value)


# ML usage:
# Statistical calculations on data.
```

## 5. Function Returning Multiple Values


```python
def min_max(numbers):
    return min(numbers), max(numbers)

values = [10, 20, 5, 40]
minimum, maximum = min_max(values)

print(minimum, maximum)


# ML usage:
# Used in data normalization and analysis.
```


## 6. Function for Model Evaluation Logic (Basic)


```python
def check_accuracy(acc):
    if acc >= 80:
        return "Good Model"
    else:
        return "Needs Improvement"

print(check_accuracy(85))


# ML usage:
# Evaluating model performance.

# Why Day 5 is Critical for ML Engineers

# ML pipelines are built using functions

# Reusability saves time and effort

# Clean structure helps in collaboration

# Functions are the base of ML libraries
```

## 1. Writing Data to a File

```python
file = open("data.txt", "w")
file.write("Machine Learning starts with data\n")
file.write("Python makes file handling easy")
file.close()
```

## 2. Reading Data from a File

```python
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

## 3. Reading File Line by Line


```python
file = open("data.txt", "r")
for line in file:
    print(line)
file.close()
```

## 4. Appending Data to a File

```python
file = open("data.txt", "a")
file.write("\nThis data is added later")
file.close()
```

## 5. Using with Statement (Recommended)

```python
with open("data.txt", "r") as file:
    data = file.read()
    print(data)
```

### 1. Installing and importing Numpy

```python
import numpy as np
```

### 2. Creating NumPy Arrays

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

### 3. Array with Zeros and Ones

```python
zeros = np.zeros(5)
ones = np.ones(5)

print(zeros)
print(ones)
```

### 4. Creating a 2D Array (Matrix)

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```

### 5. Array Shape and Size

```python
print(matrix.shape)
print(matrix.size)
```

### 6. Indexing and Slicing

```python
# arr = [10, 20, 30, 40, 50] normal array
# arr = [10, 20, 30, 40, 50] 
# numpy array
arr =np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[1:4])
```

### 7. Mathematical Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a * b)
```

### 8. Statistical Operations (ML Important)

```python
data = np.array([10, 20, 30, 40, 50])

print(np.mean(data))
print(np.max(data))
print(np.min(data))
print(np.std(data))

```

### ML Perspective Example (Feature Scaling Logic)

```python
data = np.array([10, 20, 30, 40, 50])
scaled_data = data / np.max(data)

print(scaled_data)
```
