# Day 16 – Hyperparameter Tuning and Model Optimization

## Introduction
Hyperparameter tuning is the process of selecting the best parameters for a Machine Learning model
to improve its performance.

Model optimization helps in achieving better accuracy and generalization on unseen data.

---

## What are Hyperparameters
Hyperparameters are settings that are defined before training the model.

They are not learned from data.
Instead, they control how the model learns.

Examples:
- Learning rate
- Number of neighbors (K in KNN)
- Depth of decision tree
- Number of estimators in Random Forest

---

## Difference Between Parameters and Hyperparameters

Parameters:
- Learned automatically during training
- Example: Weights in Linear Regression

Hyperparameters:
- Set manually before training
- Example: Number of trees in Random Forest

---

## Why Hyperparameter Tuning is Important
- Improves model accuracy
- Reduces overfitting
- Enhances generalization
- Makes model more stable

---

## Common Hyperparameter Tuning Techniques

### Grid Search
Tests all possible combinations of given hyperparameter values.

### Random Search
Tests random combinations instead of all possibilities.

### Cross Validation
Used along with tuning to evaluate performance properly.

---

## Model Optimization
Model optimization includes:
- Selecting best hyperparameters
- Feature selection
- Regularization
- Improving training process

---

## Conclusion
Hyperparameter tuning is essential for building high-performance Machine Learning models.
Without proper tuning, even strong algorithms may perform poorly.
